# Data Preprocessing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# 1. Convert date column to datetime
df["SETTLEMENT_DATE"] = pd.to_datetime(df["SETTLEMENT_DATE"], format='mixed', dayfirst=True)

# 2. Ensure settlement period is numeric
df["SETTLEMENT_PERIOD"] = pd.to_numeric(df["SETTLEMENT_PERIOD"])

# 3. Create full timestamp (each period = 30 minutes)
df["timestamp"] = df["SETTLEMENT_DATE"] + pd.to_timedelta(
    (df["SETTLEMENT_PERIOD"] - 1) * 30, unit="min"
)

# 4. Set timestamp as index
df = df.set_index("timestamp")

# 5. Sort by time
df = df.sort_index()

# 6. Assign demand column
demand_col = "ND"
demand = df[demand_col]

# 7. Basic dataset info
print("Data shape:", df.shape)
print("Date range:", df.index.min(), "to", df.index.max())

# 8. Demand statistics
print(demand.describe())
```

# Base Model

```python
y = series.sort_index()

# seasonal naive: t prediction = value at t-48 (same settlement period previous day)
y_pred = y.shift(48)

# Evaluate on overlapping part
valid = y.dropna().index.intersection(y_pred.dropna().index)
mae = (y.loc[valid] - y_pred.loc[valid]).abs().mean()

print("Baseline Seasonal-Naive MAE:", float(mae))

# Plot 2 days example
end = y.index.max()
start = end - pd.Timedelta(days=2)
plt.figure(figsize=(14,4))
plt.plot(y.loc[start:end], linewidth=1.2)
plt.plot(y_pred.loc[start:end], linewidth=1.2)
plt.title("Baseline Forecast Example (Seasonal Naive: yesterday same period)")
plt.legend(["Actual", "Baseline forecast"])
plt.show()
```

# LSTM Model

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error

# -------------------------------
# 1. TRAIN / TEST SPLIT
# -------------------------------
train = series[:'2023']
test  = series['2024':]

# -------------------------------
# 2. SCALE DATA
# -------------------------------
scaler = MinMaxScaler()

train_scaled = scaler.fit_transform(train.values.reshape(-1,1))
test_scaled  = scaler.transform(test.values.reshape(-1,1))

# -------------------------------
# 3. CREATE SEQUENCES (REQUIRED)
# -------------------------------
def create_sequences(data, window):
    X, y = [], []
    for i in range(len(data)-window):
        X.append(data[i:i+window])
        y.append(data[i+window])
    return np.array(X), np.array(y)

SEQ_LEN = 48

X_train, y_train = create_sequences(train_scaled, SEQ_LEN)
X_test, y_test   = create_sequences(test_scaled, SEQ_LEN)

# -------------------------------
# 4. BUILD IMPROVED MODEL
# -------------------------------
model = Sequential()

model.add(LSTM(64, return_sequences=True, input_shape=(SEQ_LEN,1)))
model.add(Dropout(0.2))

model.add(LSTM(64, return_sequences=True))
model.add(Dropout(0.2))

model.add(LSTM(32))
model.add(Dropout(0.2))

model.add(Dense(1))

model.compile(
    optimizer='adam',
    loss='mse',
    metrics=['mae']
)

model.summary()

# -------------------------------
# 5. TRAIN MODEL (SMART TRAINING)
# -------------------------------
early_stop = EarlyStopping(
    monitor='val_loss',
    patience=3,
    restore_best_weights=True
)

history = model.fit(
    X_train, y_train,
    epochs=15,
    batch_size=64,   # faster
    validation_split=0.1,
    callbacks=[early_stop],
    verbose=1
)

```

```python
# -------------------------------
# 6. PLOT TRAINING
# -------------------------------
plt.figure(figsize=(8,4))
plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Validation')
plt.legend()
plt.title("Training vs Validation Loss")
plt.show()

# -------------------------------
# 7. PREDICTIONS
# -------------------------------
pred = model.predict(X_test)

pred = scaler.inverse_transform(pred)
y_test_real = scaler.inverse_transform(y_test)

# -------------------------------
# 8. EVALUATION
# -------------------------------
mae = mean_absolute_error(y_test_real, pred)
rmse = np.sqrt(mean_squared_error(y_test_real, pred))

print("LSTM MAE:", mae)
print("LSTM RMSE:", rmse)

# -------------------------------
# 9. PLOT RESULTS
# -------------------------------
plt.figure(figsize=(12,4))
plt.plot(y_test_real[:300], label="Actual")
plt.plot(pred[:300], label="Forecast")
plt.legend()
plt.title("LSTM Forecast vs Actual")
plt.show()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.metrics import mean_absolute_error

configs = [
    {"seq":48, "units":32, "dropout":0.1},
    {"seq":48, "units":64, "dropout":0.2},
    {"seq":96, "units":64, "dropout":0.2}
]

results = []

for config in configs:
    SEQ_LEN = config["seq"]

    X_train, y_train = create_sequences(train_scaled, SEQ_LEN)
    X_test, y_test   = create_sequences(test_scaled, SEQ_LEN)

    model = Sequential()

    # Layer 1
    model.add(LSTM(config["units"], return_sequences=True, input_shape=(SEQ_LEN,1)))
    model.add(Dropout(config["dropout"]))

    # Layer 2
    model.add(LSTM(config["units"]))
    model.add(Dropout(config["dropout"]))

    # Output
    model.add(Dense(1))

    model.compile(optimizer='adam', loss='mse')

    model.fit(
        X_train, y_train,
        epochs=5,          # keep small for tuning
        batch_size=32,
        verbose=0
    )

    pred = model.predict(X_test, verbose=0)

    pred = scaler.inverse_transform(pred)
    y_test_real = scaler.inverse_transform(y_test)

    mae = mean_absolute_error(y_test_real, pred)

    results.append((config, mae))

# Print results
for r in results:
    print(r)
```

```python
from statsmodels.tsa.arima.model import ARIMA
import numpy as np

train_ts = series[:'2023']
test_ts  = series['2024':]

# Fit ARIMA
model = ARIMA(train_ts, order=(2,1,2))
model_fit = model.fit()

# Forecast
forecast = model_fit.forecast(steps=len(test_ts))

# Evaluation
mae = np.mean(np.abs(test.values - forecast.values))
rmse = np.sqrt(np.mean((test.values - forecast.values)**2))

print("MAE:", mae)
print("RMSE:", rmse)

# Plot
plt.figure(figsize=(12,4))
plt.plot(test_ts[:500], label='Actual')
plt.plot(forecast[:500], label='ARIMA Forecast')
plt.legend()
plt.title("ARIMA Forecast")
plt.show()
```

```python
import itertools

p = [1,2]
d = [1]
q = [1,2]

best_mae = float("inf")
best_order = None

for order in itertools.product(p,d,q):
    try:
        model = ARIMA(train_ts, order=order)
        model_fit = model.fit()
        forecast = model_fit.forecast(steps=len(test_ts))

        mae = np.mean(np.abs(test.values - forecast.values))

        if mae < best_mae:
            best_mae = mae
            best_order = order
    except:
        continue

print("Best ARIMA:", best_order, "MAE:", best_mae)
```
