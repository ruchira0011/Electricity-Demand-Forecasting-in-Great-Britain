```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# 1️⃣ Convert date column to datetime
df["SETTLEMENT_DATE"] = pd.to_datetime(df["SETTLEMENT_DATE"], format='mixed', dayfirst=True)

# 2️⃣ Ensure settlement period is numeric
df["SETTLEMENT_PERIOD"] = pd.to_numeric(df["SETTLEMENT_PERIOD"])

# 3️⃣ Create full timestamp (each period = 30 minutes)
df["timestamp"] = df["SETTLEMENT_DATE"] + pd.to_timedelta(
    (df["SETTLEMENT_PERIOD"] - 1) * 30, unit="min"
)

# 4️⃣ Set timestamp as index
df = df.set_index("timestamp")

# 5️⃣ Sort by time
df = df.sort_index()

# 6️⃣ Assign demand column
demand_col = "ND"
demand = df[demand_col]

# 7️⃣ Basic dataset info
print("Data shape:", df.shape)
print("Date range:", df.index.min(), "to", df.index.max())

# 8️⃣ Demand statistics
print(demand.describe())
```

# Base Model

```python
# create time features
tmp = df[[demand_col]].copy()
tmp["hour"] = tmp.index.hour
tmp["minute"] = tmp.index.minute
tmp["hhmm"] = tmp["hour"] + tmp["minute"]/60
tmp["weekday"] = tmp.index.day_name()
tmp["is_weekend"] = (tmp.index.weekday >= 5)

# Intraday average profile (all days)
intraday = tmp.groupby("hhmm")[demand_col].mean()

plt.figure(figsize=(10,4))
plt.plot(intraday.index, intraday.values)
plt.title("Average Intraday Demand Profile (All Days)")
plt.xlabel("Hour of day")
plt.ylabel("Demand")
plt.xticks(range(0,24,2))
plt.show()

# Weekday vs weekend profile
intraday_wd = tmp[tmp["is_weekend"]==False].groupby("hhmm")[demand_col].mean()
intraday_we = tmp[tmp["is_weekend"]==True].groupby("hhmm")[demand_col].mean()

plt.figure(figsize=(10,4))
plt.plot(intraday_wd.index, intraday_wd.values)
plt.plot(intraday_we.index, intraday_we.values)
plt.title("Intraday Profile: Weekday vs Weekend")
plt.xlabel("Hour of day")
plt.ylabel("Demand")
plt.xticks(range(0,24,2))
plt.legend(["Weekday", "Weekend"])
plt.show()
```

```python
by_weekday = tmp.groupby("weekday")[demand_col].mean()
# order properly
order = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
by_weekday = by_weekday.reindex(order)

plt.figure(figsize=(8,4))
plt.bar(by_weekday.index, by_weekday.values)
plt.title("Average Demand by Weekday")
plt.ylabel("Demand")
plt.xticks(rotation=30)
plt.show()
```

```python
pivot = tmp.copy()
pivot["hour"] = pivot.index.hour
pivot["weekday_num"] = pivot.index.weekday  # Mon=0
heat = pivot.groupby(["weekday_num","hour"])[demand_col].mean().unstack()

plt.figure(figsize=(12,4))
plt.imshow(heat.values, aspect="auto")
plt.title("Mean Demand Heatmap (Weekday x Hour)")
plt.xlabel("Hour of day")
plt.ylabel("Weekday (Mon→Sun)")
plt.xticks(range(0,24,2), range(0,24,2))
plt.yticks(range(0,7), ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"])
plt.colorbar(label="Demand")
plt.show()
```

```python
y = series.sort_index()

# seasonal naive: t prediction = value at t-48 (same settlement period previous day)
y_pred = y.shift(48)

# Evaluate on overlapping part
valid = y.dropna().index.intersection(y_pred.dropna().index)
mae = (y.loc[valid] - y_pred.loc[valid]).abs().mean()

print("Baseline Seasonal-Naive MAE:", float(mae))

# Plot 2 days example
end = y.index.max()
start = end - pd.Timedelta(days=2)
plt.figure(figsize=(14,4))
plt.plot(y.loc[start:end], linewidth=1.2)
plt.plot(y_pred.loc[start:end], linewidth=1.2)
plt.title("Baseline Forecast Example (Seasonal Naive: yesterday same period)")
plt.legend(["Actual", "Baseline forecast"])
plt.show()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# -------------------------------
# Scale data
# -------------------------------
scaler = MinMaxScaler()

train = series[:'2023']
test  = series['2024':]

train_scaled = scaler.fit_transform(train.values.reshape(-1,1))
test_scaled  = scaler.transform(test.values.reshape(-1,1))

# -------------------------------
# Create sequences
# -------------------------------
def create_sequences(data, window):
    X, y = [], []
    for i in range(len(data)-window):
        X.append(data[i:i+window])
        y.append(data[i+window])
    return np.array(X), np.array(y)

SEQ_LEN = 48  # 1 day

X_train, y_train = create_sequences(train_scaled, SEQ_LEN)
X_test, y_test   = create_sequences(test_scaled, SEQ_LEN)

# -------------------------------
# Build model
# -------------------------------
model = Sequential()
model.add(LSTM(64, return_sequences=True, input_shape=(SEQ_LEN,1)))
model.add(LSTM(32))
model.add(Dense(1))

model.compile(
    optimizer='adam',
    loss='mse',
    metrics=['mae']
)

# -------------------------------
# Train model
# -------------------------------
history = model.fit(
    X_train, y_train,
    epochs=5,
    batch_size=32,
    validation_split=0.1
)

# -------------------------------
# Plot training vs validation
# -------------------------------
import matplotlib.pyplot as plt

plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='validation')
plt.legend()
plt.title("Training vs Validation Loss")
plt.show()

# -------------------------------
# Predictions
# -------------------------------
pred = model.predict(X_test)

pred = scaler.inverse_transform(pred)
y_test_real = scaler.inverse_transform(y_test)

# -------------------------------
# Evaluation
# -------------------------------
mae = mean_absolute_error(y_test_real, pred)
rmse = np.sqrt(mean_squared_error(y_test_real, pred))

print("LSTM MAE:", mae)
print("LSTM RMSE:", rmse)
```

```python
# Get correct time index for test set
test_index = test.index[SEQ_LEN:]  # skip first 48 due to sequence creation

plt.figure(figsize=(12,4))
plt.plot(test_index[:500], y_test_real[:500], label='Actual')
plt.plot(test_index[:500], pred[:500], label='Forecast')
plt.legend()
plt.title("LSTM Demand Forecast (with correct time axis)")
plt.xlabel("Date")
plt.ylabel("Demand")
plt.show()
```

```python
configs = [
    {"seq":48, "units":32},
    {"seq":48, "units":64},
    {"seq":96, "units":64}
]

results = []

for config in configs:
    SEQ_LEN = config["seq"]

    X_train, y_train = create_sequences(train_scaled, SEQ_LEN)
    X_test, y_test   = create_sequences(test_scaled, SEQ_LEN)

    model = Sequential()
    model.add(LSTM(config["units"], input_shape=(SEQ_LEN,1)))
    model.add(Dense(1))

    model.compile(optimizer='adam', loss='mse')

    model.fit(X_train, y_train, epochs=3, batch_size=32, verbose=0)

    pred = model.predict(X_test)
    pred = scaler.inverse_transform(pred)
    y_test_real = scaler.inverse_transform(y_test)

    mae = mean_absolute_error(y_test_real, pred)

    results.append((config, mae))

for r in results:
    print(r)
```

```python
from statsmodels.tsa.arima.model import ARIMA
import numpy as np

train_ts = series[:'2023']
test_ts  = series['2024':]

# Fit ARIMA
model = ARIMA(train_ts, order=(2,1,2))
model_fit = model.fit()

# Forecast
forecast = model_fit.forecast(steps=len(test_ts))

# Evaluation
mae = np.mean(np.abs(test.values - forecast.values))
rmse = np.sqrt(np.mean((test.values - forecast.values)**2))

print("MAE:", mae)
print("RMSE:", rmse)

# Plot
plt.figure(figsize=(12,4))
plt.plot(test_ts[:500], label='Actual')
plt.plot(forecast[:500], label='ARIMA Forecast')
plt.legend()
plt.title("ARIMA Forecast")
plt.show()
```

```python
import itertools

p = [1,2]
d = [1]
q = [1,2]

best_mae = float("inf")
best_order = None

for order in itertools.product(p,d,q):
    try:
        model = ARIMA(train_ts, order=order)
        model_fit = model.fit()
        forecast = model_fit.forecast(steps=len(test_ts))

        mae = np.mean(np.abs(test.values - forecast.values))

        if mae < best_mae:
            best_mae = mae
            best_order = order
    except:
        continue

print("Best ARIMA:", best_order, "MAE:", best_mae)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

# -------------------------------
# Ensure correct frequency
# -------------------------------
df = df.asfreq("30min")

# Remove missing values
series = df["ND"].dropna()

# -------------------------------
# Train / Test split
# -------------------------------
train = series[:'2023']
test  = series['2024':]

print("Train size:", len(train))
print("Test size:", len(test))

# -------------------------------
# Build SARIMAX model
# -------------------------------
model = SARIMAX(
    train,
    order=(2,1,2),
    seasonal_order=(1,1,1,48),  # 48 = daily seasonality
    enforce_stationarity=False,
    enforce_invertibility=False
)

model_fit = model.fit(disp=False)

# -------------------------------
# Forecast
# -------------------------------
forecast = model_fit.forecast(steps=len(test))

# Align index (optional for plotting)
forecast.index = test.index

# -------------------------------
# Evaluation
# -------------------------------
mae = np.mean(np.abs(test.values - forecast.values))
rmse = np.sqrt(np.mean((test.values - forecast.values)**2))

print("SARIMAX MAE:", mae)
print("SARIMAX RMSE:", rmse)

# -------------------------------
# Plot
# -------------------------------
plt.figure(figsize=(12,4))
plt.plot(test[:500], label="Actual")
plt.plot(forecast[:500], label="SARIMAX Forecast")
plt.legend()
plt.title("SARIMAX Forecast vs Actual")
plt.xlabel("Date")
plt.ylabel("Demand")
plt.show()
```
